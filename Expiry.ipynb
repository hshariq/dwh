```python
import pandas as pd

# Define column names for ProductExpiryDate table
ProductExpiryDate_columns = ['EDateID', 'ExpiryYear', 
                             'ExpiryQuarter', 'ExpiryMonth', 
                             'ExpiryWeek', 'ExpiryDay']

# Create an empty DataFrame with the specified column names
df = pd.DataFrame(columns=ProductExpiryDate_columns)

# Display the empty DataFrame
df
```

```python
import random
edate = []
for i in range(0,500):
    a= random.randint(1,300)
    edate.append(a)
df['EDateID'] = edate
df
```

```python
y = []
q= []
m = []
w = []
d = []
for i in range(0,500):
    a= random.randint(2024,2026)
    y.append(a)
    q.append(random.randint(1,4))
    m.append(random.randint(1,12))
    w.append(random.randint(1,52))
    d.append(random.randint(1,31))
df['ExpiryYear'] = y
df['ExpiryQuarter'] = q
df['ExpiryMonth'] = m
df['ExpiryWeek'] = w
df['ExpiryDay'] = d
df
```

```python
df.to_csv('Data/ProductExpiryDate.csv',index=False)
```
