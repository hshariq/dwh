```python
import pandas as pd

# Define column names for CustomerLocation table
CustomerLocation_columns = ['CLocID', 'CustomerCountry', 'CustomerState', 
                            'CustomerCity', 'CustomerArea', 'CustomerZIPCode',
                            'IsDefault', 'CustomerID']

# Create an empty DataFrame with the specified column names
df = pd.DataFrame(columns=CustomerLocation_columns)

# Display the empty DataFrame
df
```

```python
ids= []
for i in range(0,25):
   ids.append(i+1)

df['CLocID'] = ids
df
```

```python
import random
count = []

for i in range(0,25):
    c = random.randint(100, 9999)
    if c<1000:
        count.append('USA')
    elif c<2000:
        count.append('Pakistan')
    else:
        count.append('Canada')

df['CustomerCountry'] = count
```

```python
import random
state = []

for i in range(0,25):
    c = random.randint(100, 9999)
    if df['CustomerCountry'][i] == 'USA':   
        if c<1000:
            state.append('New York')
        elif c<2000:
            state.append('California')
        else:
            state.append('Texas')
    elif df['CustomerCountry'][i] == 'Pakistan':
        if c<1000:
            state.append('Punjab')
        elif c<2000:
            state.append('Sindh')
        else:
            state.append('KPK')
    else:
        if c<1000:
            state.append('Ontario')
        elif c<2000:
            state.append('Quebec')
        else:
            state.append('British Columbia')
    
df['CustomerState'] = state
df
```

```python
import random
city = []

for i in range(0,25):
    c = random.randint(100, 9999)
    if df['CustomerState'][i] == 'British Columbia':   
        if c<1000:
            city.append('Vancouver')
        elif c<2000:
            city.append('Victoria')
        else:
            city.append('Kelowna')
    elif df['CustomerState'][i] == 'Ontario':
        if c<1000:
            city.append('Toronto')
        elif c<2000:
            city.append('Ottawa')
        else:
            city.append('Hamilton')
    elif df['CustomerState'][i] == 'KPK':
            city.append('Peshawar')
    elif df['CustomerState'][i] == 'Quebec':
        if c<1000:
            city.append('Montreal')
        elif c<2000:
            city.append('Quebec City')
        else:
            city.append('Sherbrooke')
    elif df['CustomerState'][i] == 'Sindh':
        if c<1000:
            city.append('Karachi')
        elif c<2000:
            city.append('Hyderabad')
        else:
            city.append('Sukkur')
            
df['CustomerCity'] = city
df
```
