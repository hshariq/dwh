```python
import datetime
dates = []

# Define a function to calculate the shipping date
def calculate_shipping_date(order_date, shipping_mode):
    # Convert the order date string to a datetime object
    order_date = datetime.datetime.strptime(order_date, "%d-%m-%Y")
    
    # Calculate the number of days to add based on the shipping mode
    if shipping_mode == 'Air':
        days_to_add = 3
    elif shipping_mode == 'Road':
        days_to_add = 5
    elif shipping_mode == 'Sea':
        days_to_add = 10
    else:
        days_to_add = 0
    
    # Calculate the shipping date by adding the number of days to the order date
    shipping_date = order_date + datetime.timedelta(days=days_to_add)
    
    # Return the shipping date as a string in the format "dd-mm-yyyy"
    return shipping_date.strftime("%d-%m-%Y")

# Apply the function to each row in the DataFrame
for index, row in df.iterrows():
    order_date = row['OrderDate']
    shipping_mode = row['ShippingMode']
    dates.append(calculate_shipping_date(order_date, shipping_mode))

# Ensure that the 'dates' list has the same length as the DataFrame
if len(dates) == len(df):
    df['ShippingDate'] = dates
else:
    print("Error: Length of 'dates' list does not match length of DataFrame.")

for index, row in df.iterrows():
    order_date = row['OrderDate']
    shipping_mode = row['ShippingMode']
    calculate_shipping_date(order_date, shipping_mode)

df['ShippingDate'] = dates
```
