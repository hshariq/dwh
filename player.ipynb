```python
import random 
names = []
genders = []
address = []
passwords = []
for i in range(0,15):
    num = random.randint(1,2)
    if (num == 1):
        names.append(fake.first_name_male() +" "+ fake.last_name())
        genders.append("M")
    else:
        names.append(fake.first_name_female() +" "+ fake.last_name())
        genders.append("F")
    # temp = fake.address()
    # temp = temp.replace("\n"," ")
    # temp = temp.replace(",", "")
    # address.append(temp)
    passwords.append(fake.password())

df['ManagerName'] = names
df['ManagerGender'] = genders
# df['player_address'] = address
# df['player_password'] = passwords

df
```

```python
import random
import string

def generate_email(name):
    domain = "@gmail.com" 
    name = name.replace(' ','')
    random_part = ''.join(random.choices(string.ascii_lowercase + string.digits, k=3))
    return name.lower() + '.' + random_part + domain


df['ManagerEmail'] = df['ManagerName'].apply(generate_email)

df
```
