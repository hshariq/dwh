```python
import pandas as pd

# Define column names for Product table
Product_columns = ['ProductID', 'ProductName', 
                   'ProductCategory', 'ProductDescription',
                   'ProductColor', 'ProductSize', 
                   'ProductWeight', 'InStock', 'EDateID']

# Create an empty DataFrame with the specified column names
df = pd.DataFrame(columns=Product_columns)

# Display the empty DataFrame
df
```

```python
import random
q= []
for i in range(0,500):
    a= random.randint(1,200)
    q.append(a)
        
df['ProductID'] = q
df
```

```python
df['ProductName'] = df['ProductID'].map(lambda x: "Product" + str(x))
df
```

```python
import random
proCat = []

for i in range(0,500):
    c = random.randint(100, 9999)
    if c<1000:
        proCat.append('Dessert')  
    elif c<2000:
        proCat.append('Beverage')
    else:
        proCat.append('Main_Course')

df['ProductCategory'] = proCat
df
```

```python
df['ProductDescription'] = df['ProductCategory'].map(lambda x: "This is a " + str(x))
df
```

```python
# Create a list of random colors
colors = ['Red', 'Blue', 'Green', 'Yellow', 'White', 'Black', 'Purple', 'Orange', 'Pink', 'Brown']
rang=[]
for i in range(0,500):
    c = random.choice(colors)
    rang.append(c)
df['ProductColor'] = rang
df
```

```python
# Create a list of random sizes
sizes = ['S', 'M', 'L']
s=[]
for i in range(0,500):
    c = random.choice(sizes)
    s.append(c)
df['ProductSize'] = s
df
```

```python
# Create a list of random weights
weights = [0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]
w=[]
for i in range(0,500):
    c = random.choice(weights)
    w.append(c)
df['ProductWeight'] = w
df
```

```python
# Create a list of random in stock values
inStock = [True, False]
ins=[]
for i in range(0,500):
    c = random.choice(inStock)
    ins.append(c)
df['InStock'] = ins
df
```

```python
# Create a list of random EDateID values
edate = []
for i in range(0,500):
    a= random.randint(1,300)
    edate.append(a)
df['EDateID'] = edate
df
```

```python
df.to_csv('Data/Product.csv', index=False)
```
