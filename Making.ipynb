```python
#random for QuantityProduced but make sure for each row QuantityProduced is greater than QuantityUsed
ids= []
for i in range(0,450):
    quantity_used = df.loc[i, 'QuantityUsed']
    quantity_produced = random.randint(quantity_used + 1, 300)
    ids.append(quantity_produced)
df['QuantityProduced'] = ids
```
