```python
import pandas as pd

# Define column names for Inventory table
Inventory_columns = ['InventoryID', 'InventoryName', 
                     'InventoryRackNumber', 'InventoryBatchNumber', 
                     'Inventory_Initial_Quantity_Ordered', 
                     'InventoryQuantityUsed',
                     'InventoryQuantityinTransit', 
                     'InventoryQuantityinStock',
                     'InventoryReorderThreshold', 
                     'InventoryDamagedQuantity', 
                     'InventoryStockStatus', 'InventoryType', 
                     'WarehouseID', 'RawMaterialID']

# Create an empty DataFrame with the specified column names
df = pd.DataFrame(columns=Inventory_columns)

# Display the empty DataFrame
df
```

```python
import random
q= []
for i in range(1,501):
    q.append(i)
        
df['InventoryID'] = q
df
```

```python
df['InventoryName'] = df['InventoryID'].map(lambda x: "Inv" + str(x))
df
```

```python
#randomly add InventoryRackNumber
df['InventoryRackNumber'] = random.sample(range(1, 1000), 500)
df
```

```python
#randomly add InventoryRackNumber
df['InventoryBatchNumber'] = random.sample(range(1, 1500), 500)
df
```

```python
# randomly add InventoryRackNumber
df['Inventory_Initial_Quantity_Ordered'] = random.choices(range(100, 300), k=len(df))
df
```

```python
df['InventoryQuantityUsed'] = random.choices(range(1, 100), k=len(df))
df
```

```python
# randomly add InventoryRackNumber
df['InventoryQuantityinTransit'] = random.choices(range(20, 60), k=len(df))
df
```

```python
# randomly add InventoryRackNumber
df['InventoryDamagedQuantity'] = random.choices(range(5, 10), k=len(df))
df
```

```python
#qunaity orderd - qunaity used is InventoryQuantityinStock
df['InventoryQuantityinStock'] = df['Inventory_Initial_Quantity_Ordered'] - df['InventoryQuantityUsed']
df
```

```python
ids= []
for i in range(0,500):
   ids.append(random.choice(range(0, 15)))

df['WarehouseID'] = ids
df
```

```python
ids= []
for i in range(0,500):
   ids.append(random.choice(range(0, 65)))

df['RawMaterialID'] = ids
df
```

```python
#randomly add InventoryStockStatus
df['InventoryStockStatus'] = random.choices(['In Stock', 'Out of Stock', 'In Transit', 'Damaged'], k=len(df))
df
```

```python
#randomly add InventoryType
df['InventoryType'] = random.choices(['Raw Material', 'Work in Progress', 'Finished Goods'], k=len(df))
```

```python
df['InventoryReorderThreshold']= 10
```

```python
df.to_csv("Data/Inventory.csv")
```
