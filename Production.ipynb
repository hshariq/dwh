```python
import pandas as pd

# Define column names for Production table
Production_columns = ['ProductionID', 'ProductionStartDate',
                      'ProductionEndDate', 'Status', 'WarehouseID',
                      'PManagerID']

# Create an empty DataFrame with the specified column names
df = pd.DataFrame(columns=Production_columns)

# Display the empty DataFrame
df
```

```python
import pandas as pd
from faker import Faker

fake = Faker(use_weighting=True,locale='en_US')
```

```python
ids= []
for i in range(0,500):
   ids.append(i+1)

df['ProductionID'] = ids
df
```

```python
#start date
start_date = []
for _ in range(500):
    start_date.append(fake.date_between(start_date='-1y', end_date='today'))
df['ProductionStartDate'] = start_date
df
```

```python
#end date a few days later then start date
end_date = []
for _ in range(500):
    end_date.append(fake.date_between(start_date=df['ProductionStartDate'][_], end_date='+1y'))
df['ProductionEndDate'] = end_date
df
```

```python
#status ranomly
status = []
for _ in range(500):
    status.append(fake.random_element(elements=('Active', 'Inactive')))
df['Status'] = status
df
```

```python
#warehouse id 1-15 randomly
warehouse_id = []
for _ in range(500):
    warehouse_id.append(fake.random_int(min=1, max=15))
df['WarehouseID'] = warehouse_id
df
```

```python
#pmanagerID randomly 1-10
pmanager_id = []
for _ in range(500):
    pmanager_id.append(fake.random_int(min=1, max=10))
df['PManagerID'] = pmanager_id
df
```

```python
df.to_csv('Data/Production.csv',index=False)
```
