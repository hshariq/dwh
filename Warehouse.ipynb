```python
import random
count = []

for i in range(0,15):
    c = random.randint(100, 9999)
    if c<1000:
        count.append('USA')
    elif c<2000:
        count.append('Pakistan')
    else:
        count.append('Canada')

df['WarehouseCountry'] = count
df
```

```python
import random
state = []

for i in range(0,15):
    c = random.randint(100, 9999)
    if df['WarehouseCountry'][i] == 'USA':   
        if c<1000:
            state.append('New York')
        elif c<2000:
            state.append('California')
        else:
            state.append('Texas')
    elif df['WarehouseCountry'][i] == 'Pakistan':
        if c<1000:
            state.append('Punjab')
        elif c<2000:
            state.append('Sindh')
        else:
            state.append('KPK')
    else:
        if c<1000:
            state.append('Ontario')
        elif c<2000:
            state.append('Quebec')
        else:
            state.append('British Columbia')
    
df['WarehouseState'] = state
df
```
