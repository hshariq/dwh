```python
import pandas as pd

# Define column names for ProductionManager table
ProductionManager_columns = ['PManagerID', 'PManagerName', 'PManagerDOB', 
                             'PManagerAge', 'PManagerGender', 'PManagerEmail', 
                             'PManagerPhoneNumber', 'PManagerCreatedDate', 'IsActive']

# Create an empty DataFrame with the specified column names
df = pd.DataFrame(columns=ProductionManager_columns)

# Display the empty DataFrame
df
```

```python
import pandas as pd
from faker import Faker

fake = Faker(use_weighting=True,locale='en_US')
```

```python
ids= []
for i in range(0,10):
   ids.append(i+1)

df['PManagerID'] = ids
df
```

```python
import random 
names = []
genders = []
address = []
passwords = []
for i in range(0,10):
    num = random.randint(1,2)
    if (num == 1):
        names.append(fake.first_name_male() +" "+ fake.last_name())
        genders.append("M")
    else:
        names.append(fake.first_name_female() +" "+ fake.last_name())
        genders.append("F")
    # temp = fake.address()
    # temp = temp.replace("\n"," ")
    # temp = temp.replace(",", "")
    # address.append(temp)
    passwords.append(fake.password())

df['PManagerName'] = names
df['PManagerGender'] = genders
# df['player_address'] = address
# df['player_password'] = passwords

df
```

```python
import random
import string

def generate_email(name):
    domain = "@gmail.com" 
    name = name.replace(' ','')
    random_part = ''.join(random.choices(string.ascii_lowercase + string.digits, k=3))
    return name.lower() + '.' + random_part + domain


df['PManagerEmail'] = df['PManagerName'].apply(generate_email)

df
```

```python
import random

dates = []
for i in range(0,10):
    mm = random.randint(1,12)
    yy = random.randint(1995,2001)
    dd = random.randint(1,28)
    dates.append(f"{dd}-{mm}-{yy}")

df['PManagerDOB'] = dates

df
    

```

```python
import random

dates = []
for i in range(0,10):
    mm = random.randint(1,12)
    yy = random.randint(2015,2024)
    dd = random.randint(1,28)
    dates.append(f"{dd}-{mm}-{yy}")

df['PManagerCreatedDate'] = dates

df
    

```

```python
import random
import datetime

# Generate random dates
dates = []
current_date = datetime.datetime.now()

for i in range(0, 10):
    dobyear = df.loc[i, 'PManagerDOB']
    dyear= int(dobyear.split("-")[2])    
    yy = current_date.year
    age = yy - dyear    
    dates.append(age)

df['PManagerAge'] = dates

df
```

```python
import random
phone_number = []

for i in range(0,10):
    area_code = random.randint(100, 999)
    prefix = random.randint(100, 999)
    line_number = random.randint(1000, 9999)
    phone_number.append(f"{area_code}-{prefix}-{line_number}")

df['PManagerPhoneNumber'] = phone_number
df
```

```python
#randomly assign IsActive as True or False
is_active = []
for i in range(0,10):
    is_active.append(random.choice([True, False]))
df['IsActive'] = is_active
df
```
