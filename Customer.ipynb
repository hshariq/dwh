```python
import pandas as pd
from faker import Faker

fake = Faker(use_weighting=True,locale='en_US')

```

```python
import pandas as pd

# Define column names for Customers table
Customers_columns = ['CustomerID', 'CustomerName', 'CustomerEmail', 
                     'Customernumber', 'CustomerJoin', 'CustomerLife',
                     'CustomerDOB', 'CustomerAge', 'CustomerGender', 
                     'IsActive', 'CustomerLastPurchaseDate']

# Create an empty DataFrame with the specified column names
df = pd.DataFrame(columns=Customers_columns)

# Display the empty DataFrame
df
```

```python
ids= []
for i in range(0,350):
   ids.append(i+1)

df['CustomerID'] = ids
df
```

```python
import random 
names = []
genders = []
address = []
passwords = []
for i in range(0,350):
    num = random.randint(1,2)
    if (num == 1):
        names.append(fake.first_name_male() +" "+ fake.last_name())
        genders.append("M")
    else:
        names.append(fake.first_name_female() +" "+ fake.last_name())
        genders.append("F")
    # temp = fake.address()
    # temp = temp.replace("\n"," ")
    # temp = temp.replace(",", "")
    # address.append(temp)
    passwords.append(fake.password())

df['CustomerName'] = names
df['CustomerGender'] = genders
# df['player_address'] = address
# df['player_password'] = passwords

df
```

```python
import random
import string

def generate_email(name):
    domain = "@gmail.com" 
    name = name.replace(' ','')
    random_part = ''.join(random.choices(string.ascii_lowercase + string.digits, k=3))
    return name.lower() + '.' + random_part + domain


df['CustomerEmail'] = df['CustomerName'].apply(generate_email)

df
```

```python
import random

dates = []
for i in range(0,350):
    mm = random.randint(1,12)
    yy = random.randint(1995,2017)
    dd = random.randint(1,28)
    dates.append(f"{dd}-{mm}-{yy}")

df['CustomerDOB'] = dates

df
    

```

```python
from datetime import datetime

current_year = datetime.now().year
df['CustomerLife'] = current_year - pd.to_datetime(df['CustomerJoin'], dayfirst=True).dt.year
```
