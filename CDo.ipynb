```python
import pandas as pd

# Define column names for CustomerDeliveryOrder table
CustomerDeliveryOrder_columns = ['DeliveryID', 'DeliveredDate',
                                 'DeliveryDuration', 'DeliveryStatus',
                                 'QuantityDelivered', 'PaymentDate',
                                 'PaymentMethod', 'PaymentStatus',
                                 'IsOrderCompleted', 'OrderID',
                                 'CLocID']

# Create an empty DataFrame with the specified column names
df = pd.DataFrame(columns=CustomerDeliveryOrder_columns)

# Display the empty DataFrame
print(df)
```

```python
ids= []
import random
for i in range(0,500):
   a=random.randint(1,200) #jitne order utne deliveries
   ids.append(a)
df['DeliveryID'] = ids
df
```

```python
q= []
import random
for i in range(0,500):
    a= random.randint(1,15)
    if a<5:
        q.append(3)
    elif a<10:
        q.append(5)
    else:
        q.append(10)
        
df['DeliveryDuration'] = q
df
```

```python
import random

dates = []
for i in range(0,500):
    mm = random.randint(1,12)
    yy = random.randint(2014,2024)
    dd = random.randint(1,28)
    dates.append(f"{dd}-{mm}-{yy}")

df['DeliveredDate'] = dates

df
    

```

```python
q= []
for i in range(0,500):
    a= random.randint(1,15)
    if a<7:
        q.append('Shipped')
    else:
        q.append('Delivered')

df['DeliveryStatus'] = q
df
```

```python
import random
q= []
for i in range(0,500):
    a= random.randint(1,15)
    q.append(a)

df['QuantityDelivered'] = q
df
```

```python
from datetime import datetime, timedelta
# Convert the delivered dates to datetime objects
delivered_dates = [datetime.strptime(date, '%d-%m-%Y') for date in df['DeliveredDate']]

# Generate payment dates ±10 days from the delivered dates
payment_dates = [date + timedelta(days=random.randint(-10, 10)) for date in delivered_dates]

# Convert the payment dates back to the desired format
payment_dates = [date.strftime('%d-%m-%Y') for date in payment_dates]

# Assign the payment dates to the 'PaymentDate' column in the dataframe
df['PaymentDate'] = payment_dates

# Display the updated dataframe
df
```

```python
p=[]
for i in range(0,500):
    p.append(random.choice(['Cash','Card','Paypal']))
df['PaymentMethod'] = p
```

```python
q= []
for i in range(0,500):
    a= random.randint(1,15)
    if a<7:
        q.append('Paid')
    else:
        q.append('Pending')

df['PaymentStatus'] = q
df
```
